7.What is the total amount Pizza store lost in 2023 due to refunds on late deliveries?

```python
dic={}
for i,j in zip(products['item_id'],products['Price']):
    dic[i]=j
```

```python
order_items['items_ordered'] = order_items['items_ordered'].str.replace('[',"",regex= False).str.replace("]","",regex= False).str.replace("'",'',regex= False)
```

```python
ordered_amount=[]
for items in order_items['items_ordered']:
    amount=0
    for item in items.split(','):
        amount+= dic[(item.strip())]
    ordered_amount.append(amount)
```

```python
order_items['order_amount']=ordered_amount
```

```python
refu=avg_time[avg_time['delivery_time']>30]
```

```python
refunded_order_amounts=pd.merge(refu,order_items,how='inner',on='order_id')[['order_id','order_placed_at','order_amount']]
```

```python
refunded_order_amounts[refunded_order_amounts['order_placed_at'].dt.year==2023]['order_amount'].sum()
```

8.In which year Pizza store has lost maximum amount due to refunds on late deliveries?

```python
refunded_order_amounts['ordered_year']=refunded_order_amounts['order_placed_at'].dt.year
```

```python
refunded_order_amounts.groupby(['ordered_year'])[['order_amount']].sum().sort_values(by='order_amount',ascending=False).head(1)
```
